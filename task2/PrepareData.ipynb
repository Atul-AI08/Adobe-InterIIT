```python
df = pd.read_csv("output_final.csv")

train = []
for i, row in df.iterrows():
    artifacts = categories[row["label"]]
    major_artifacts = categories["major"]

    train.append(
        {
            "messages": [
                {
                    "content": f"<image><image>{prompt}\n[{', '.join(major_artifacts)}{', '.join(artifacts)}]",
                    "role": "user",
                },
                {"content": row["response"], "role": "assistant"},
            ],
            "images": [
                f"../task2_images/{row['image_path']}",
                f"../gradcam_images/{row['image_path']}",
            ],
        }
    )

with open("train.json", "w") as f:
    json.dump(train, f, indent=2)
```
