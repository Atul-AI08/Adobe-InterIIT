```python
from transformers import pipeline
import os
from PIL import Image
import pandas as pd
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
fake_path = "/content/CIFAKE_final/test/FAKE"
real_path = "/content/CIFAKE_final/test/REAL"
# final_path = "/content/perturbed_images_32_ViTpredictions"
```

```python
pipe = pipeline("image-classification", model="dima806/ai_vs_real_image_detection", device = device)
```

```python
i = 0
for filename in os.listdir(real_path):
  i += 1
i
```

```python
i_r = 0
countr = 0

for filename in os.listdir(real_path):
  path = os.path.join(real_path, filename)
  if(path == "/content/perturbed_images_32/.ipynb_checkpoints"):
    continue
  else:
    if(countr%500 == 0):
      print(f"On count: {countr}")
    countr += 1
    img = Image.open(path)
    label = pipe(img)[0]['label']
    if(label == 'REAL'):
      i_r += 1

i_f = 0
countf = 0
for filename in os.listdir(fake_path):
  path = os.path.join(fake_path, filename)
  if(path == "/content/perturbed_images_32/.ipynb_checkpoints"):
    continue
  else:
    if(countf%500 == 0):
      print(f"On count: {countf}")
    countf += 1
    img = Image.open(path)
    label = pipe(img)[0]['label']
    if(label == 'FAKE'):
      i_f += 1
```

```python
# Real
print(i_r, countr, 100*i_r/countr)
# Fake
print(i_f, countf, 100*i_f/countf)
```
